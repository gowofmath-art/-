## 說明：S的非空子集 

當然可以用數學的方式直接列出全部的非空子集合。但這邊我想嘗試使用程式的方式列出來。

核心思想：
集合有 n 個元素，每個元素都有「選 / 不選」兩種狀態。由排列組合，共有 2^n 種子集合。

而二進位剛好就是一串 0 / 1。  
因此可以用 0 到 2^n - 1 的所有二進位數，完整表示所有子集合。  

數學的思考：  
用 i 的二進位表示法，決定哪些元素要被選入子集合。  
每一個 bit 對應集合中的一個元素。  
bit = 1 → 選  
bit = 0 → 不選  

說明如下：  
針對A={a1,a2,a3}，每個子集合其實就是一個 0/1 序列：  
000 → {}  
001 → {a1}  
010 → {a2}  
011 → {a1, a2}  
100 → {a3}  
101 → {a1, a3}  
110 → {a2, a3}  
111 → {a1, a2, a3}  
可以想像：子集合⟷二進位數，這就是一個 bijection，雙射。  
每個子集合唯一對應一個二進位數，每個二進位數也唯一對應一個子集合。

所以這個演算法真正的數學核心是：  
P(A)≅{0,1}^n
也就是：一個 n 元集合的冪集合，可以視為所有長度為 n 的 0/1 向量。

## 問題 2.5.4 問題 ID：011 ⭐⭐⭐  
給定正整數 $N$。請製作程式，按照從小到大的順序輸出 $N$ 以下的質數。  
該問題可以利用埃拉托斯特尼篩法（➡️ 第 4.4 節），以計算複雜度 $O(N \log \log N)$ 來解決，但此處可容許到 $O(N^2)$。

```python
# 1. 法一：最基本的方法。兩個函式，一個判斷質數，一個只要是質數就蒐集下來。

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2,int(n**(0.5))+1):
        if n % i == 0:
            return False
    return True

def prime1(n):
    A = []
    for i in range(2, n + 1):
        if is_prime(i):
            A.append(i)
    return A


print(prime1(100))
```

```python
# 另外解：法一 （雙層迴圈）
def prime1_mixed(n):
    A = []
    for i in range(2, n + 1):
        is_prime = True
        if i <= 1:
            is_prime = False
        else:
            for j in range(2, int(i**0.5) + 1):
                if i % j == 0:
                    is_prime = False
                    break
        if is_prime:
            A.append(i)
    return A
```

```python
# 2. 法二：埃拉托斯特尼篩法。

def sieve_of_eratosthenes(n):
    if n < 2:
        return []
    # 1. 初始化：先假設 0 到 n 全都是質數 (True)
    is_prime = [True] * (n + 1)
    # 0 和 1 不是質數，手動劃掉
    is_prime[0] = is_prime[1] = False

    # 2. 開始篩選：只需要檢查到根號 n 即可
    for i in range(2, int(n**0.5) + 1):
        if is_prime[i]:  # 如果 i 還沒被劃掉，它就是質數！
            # 3. 關鍵：把 i 的倍數全部劃掉
            # 從 i*i 開始（因為比 i*i 小的倍數早就被更小的質數劃掉了）
            # 每次跳 i 格 (i*i, i*i + i, i*i + 2i...)，通通改成 False
            for j in range(i * i, n + 1, i):
                is_prime[j] = False

    # 4. 收集結果：把所有保持為 True 的索引（數字）拿出來
    primes = [num for num in range(2, n + 1) if is_prime[num]]
    return primes

# 測試
print(sieve_of_eratosthenes(50))


```

## 說明：找質數

## 1. 法一：基本方法，一個判斷質數，一個只要是質數就蒐集下來。    
時間複雜度：是 $O(N \sqrt{N})$  
外層迴圈：從 $2$ 走到 $N$，總共跑了 $N$ 次。  
內層檢查 (is_prime)：多虧了 int(n(0.5))，每次檢查一個數 $i$，最多只需要除到 $\sqrt{i}$ 次。  
當 $i$ 接近 $N$ 時，最多就是檢查 $\sqrt{N}$ 次。

## 2. 法二：埃拉托斯特尼篩法。
核心想法：  
初始狀態：假定一開始 $2$ 到 $N$ 的所有數字都是質數（開關全部設為 True）。  
索引即數字：為了讓陣列的「索引（Index）」直接對應到「數字」，我們會建立一個長度為 $N+1$ 的清單。  
例如 is_prime[5] 就代表數字 $5$ 是不是質數。  
劃掉的動作：當我們要劃掉 $3$ 的倍數時，就是把 is_prime[6], is_prime[9], is_prime[12]... 的值全部改成 False。  
是個非常偉大的做法。  實踐了真正的數學實作，更是大大降低時間複雜度，只是程式碼當然不是那麼簡單的構造。

至於為何時間複雜度可以降為 $O(N \log \log N)$ ?

## 問題 2.5.7 ⭐⭐⭐

當以下程式運行結束時，cnt 的值是多少？另外，請用 $O$ 表示此程式的計算複雜度。

```python
'''
int cnt = 0;
for (int i = 1; i <= N; i++) {
    for (int j = i + 1; j <= N; j++) {
        cnt++;
    }
}
'''
```

```python
# 可參考以下：

def count(n):
    c = 0
    for i in range(1,n+1):
        for j in range(i+1 , n+1):
            c = c+1
    return c   
print(count(11))    
```
