### 問題 3.5.1 ★
1. 將出現正面機率為 $50\%$ 的硬幣扔 $10000$ 次時，正面出現的次數可以近似於平均 $\mu$、標準差 $\sigma$ 的常態分布。請求出 $\mu$ 和 $\sigma$ 的值。  

2. 出現了正面的次數為 $4900$ 次以上 $5100$ 次以下的機率是多少？  

3. 太郎扔了同樣的硬幣 $10000$ 次， $5800$ 次出現了正面。這個硬幣出現正面的機率可以視為 $50\%$ 嗎？  

## 說明與解答
1. $$\mu = n \cdot p = 10000 \cdot 0.5 = 5000$$
$$\sigma = \sqrt{n \cdot p \cdot (1-p)} = \sqrt{10000 \cdot 0.5 \cdot 0.5} = \sqrt{2500} = 50$$


2. 首先計算 $Z$ 分數（標準化）：把邊界值 $4900$ 和 $5100$ 帶入 $Z = \frac{X - \mu}{\sigma}$。得到Z=2 , Z=-2 這樣的區間。  
接著查表計算機率：求出 $P(4900 \le X \le 5100)$ 對應的標準常態分布機率。  其實就是兩個標準差，所以其對應的機率為 95%

3. 計算它的 $Z$ 分數：$$Z = \frac{5800 - 5000}{50} = \frac{800}{50} = 16$$

在常態分布中，$Z = 16$ 代表這個觀測值偏離了平均值整整 $16$ 個標準差！  
我們剛剛才聊到，偏離 $2$ 個標準差（$Z = \pm 2$）以外的機率就只剩不到 $5\%$ 了。偏離 $16$ 個標準差在統計學上出現的機率趨近於 $0$  
因為 $5800$ 次遠遠超出了合理的隨機誤差範圍（拒絕虛無假設），所以絕對不能視為 $50\%$。這枚硬幣有極大的機率是不公正的
