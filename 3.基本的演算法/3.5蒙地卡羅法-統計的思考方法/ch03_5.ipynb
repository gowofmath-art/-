### 問題 3.5.1 ★
1. 將出現正面機率為 $50\%$ 的硬幣扔 $10000$ 次時，正面出現的次數可以近似於平均 $\mu$、標準差 $\sigma$ 的常態分布。請求出 $\mu$ 和 $\sigma$ 的值。  

2. 出現了正面的次數為 $4900$ 次以上 $5100$ 次以下的機率是多少？  

3. 太郎扔了同樣的硬幣 $10000$ 次， $5800$ 次出現了正面。這個硬幣出現正面的機率可以視為 $50\%$ 嗎？  

## 說明與解答
1. $$\mu = n \cdot p = 10000 \cdot 0.5 = 5000$$
$$\sigma = \sqrt{n \cdot p \cdot (1-p)} = \sqrt{10000 \cdot 0.5 \cdot 0.5} = \sqrt{2500} = 50$$


2. 首先計算 $Z$ 分數（標準化）：把邊界值 $4900$ 和 $5100$ 帶入 $Z = \frac{X - \mu}{\sigma}$。得到Z=2 , Z=-2 這樣的區間。  
接著查表計算機率：求出 $P(4900 \le X \le 5100)$ 對應的標準常態分布機率。  其實就是兩個標準差，所以其對應的機率為 95%

3. 計算它的 $Z$ 分數：$$Z = \frac{5800 - 5000}{50} = \frac{800}{50} = 16$$

在常態分布中，$Z = 16$ 代表這個觀測值偏離了平均值整整 $16$ 個標準差！  
我們剛剛才聊到，偏離 $2$ 個標準差（$Z = \pm 2$）以外的機率就只剩不到 $5\%$ 了。偏離 $16$ 個標準差在統計學上出現的機率趨近於 $0$  
因為 $5800$ 次遠遠超出了合理的隨機誤差範圍（拒絕虛無假設），所以絕對不能視為 $50\%$。這枚硬幣有極大的機率是不公正的


### 問題 3.5.2 ★★★
有以座標 $(3,3)$ 為中心且半徑為 $3$ 的圓，及以座標 $(3,7)$ 為中心且半徑為 $2$ 的圓。請就此回答以下問題。  

1. 在 $0 \le x < 6$ , $0 \le y < 9$ 的長方形區域中隨機放置 $100$ 萬次點。這之中有多少個點會被兩個圓中的至少一個包含？  
2. 請利用 1. 的結果，求出被兩個圓中至少一個所包含的部分的面積。  

```python
# 使用程式碼 實作蒙地卡羅法
# if i % 100000 == 0:
    # print(x,y)
# 用意是 每10萬次才print一次，觀察進度又可以讓程式跑很快。

import random
# 第一題：
def Monte_Carlo():
    n = 1000000
    m = 0
    for i in range(n):
        x = random.uniform(0,6)
        y = random.uniform(0,9)
        if i % 100000 == 0:
            print(x,y)
        if ((x-3)**2+(y-7)**2 <= 4) or ((x-3)**2+(y-3)**2 <= 9) :
            m = m + 1
    return m
total = Monte_Carlo()
print (f"1.落在兩圓內的個數:{total}")

# 第二題：
print (f"2.估計兩圓聯集的面積:{54*total/1000000}")

```

## 說明  

兩小題的步驟如下：  
1-1. 在 $0 \le x < 6$ , $0 \le y < 9$ 的長方形區域中隨機放置 $100$ 萬次點。  
1-2. 在「以座標 $(3,3)$ 為中心且半徑為 $3$ 的圓，及以座標 $(3,7)$ 為中心且半徑為 $2$ 的圓」之中的點設為m個。m即為所求。  

估計面積  
2-1. 6*9為全體長方形面積，$\frac{m}{1000000}$ 為 此圖形在整個長方形中的比例。  
2-2. 將長方形面積*比例即是所求。  


### 純數學解析（排容原理與幾何解）

### 1. 確認目標與基本條件
本題目標為求兩圓之聯集面積，依據**排容原理**：
$$\text{面積}(A \cup B) = \text{面積}(A) + \text{面積}(B) - \text{面積}(A \cap B)$$

*   **圓 $A$**：圓心 $C_1(3,3)$，半徑 $r_1=3$ $\implies \text{面積}(A) = 9\pi$
*   **圓 $B$**：圓心 $C_2(3,7)$，半徑 $r_2=2$ $\implies \text{面積}(B) = 4\pi$
*   **圓心距 $d$**：$\sqrt{(3-3)^2 + (7-3)^2} = 4$
    *   因 $r_1 + r_2 = 5 > 4$，兩圓必相交，核心在於求解交集面積 $\text{面積}(A \cap B)$。

---

### 2. 利用餘弦定理求「圓心角」
設兩圓交點為 $P, Q$。專注於 $\triangle P C_1 C_2$（三邊長為 $3, 2, 4$）：

*   **圓 $A$ 的圓心角 $\theta_1$**：
    $$\cos\left(\frac{\theta_1}{2}\right) = \frac{3^2 + 4^2 - 2^2}{2 \cdot 3 \cdot 4} = \frac{21}{24} = \frac{7}{8} \implies \theta_1 = 2 \arccos\left(\frac{7}{8}\right)$$
*   **圓 $B$ 的圓心角 $\theta_2$**：
    $$\cos\left(\frac{\theta_2}{2}\right) = \frac{2^2 + 4^2 - 3^2}{2 \cdot 2 \cdot 4} = \frac{11}{16} \implies \theta_2 = 2 \arccos\left(\frac{11}{16}\right)$$

---

### 3. 計算弓形面積
交集區域由兩個圓的弓形組成，單一弓形面積公式為 $\frac{1}{2}r^2(\theta - \sin\theta)$：
$$\text{面積}(A \cap B) = \frac{9}{2}(\theta_1 - \sin\theta_1) + 2(\theta_2 - \sin\theta_2)$$

---

### 4. 精確解析解 (Analytical Solution)
將各部分帶回排容公式，得到該區域的精確數學宣告：
$$\text{總面積} = 13\pi - \left[ \frac{9}{2}(\theta_1 - \sin\theta_1) + 2(\theta_2 - \sin\theta_2) \right]$$

*   **數值逼近**：
    *   $\theta_1 \approx 1.0039 \text{ rad}$, $\sin\theta_1 \approx 0.8436$
    *   $\theta_2 \approx 1.6207 \text{ rad}$, $\sin\theta_2 \approx 0.9987$
    *   $\text{面積}(A \cap B) \approx 0.7213 + 1.2440 = 1.9653$
*   **最終面積**：$\approx 13\pi - 1.9653 \approx 38.8754$

可觀察，與蒙地卡羅法預測得非常接近。這就是純幾何作法求出的理論解與統計方法逼近的近似解，兩者差異甚微！
